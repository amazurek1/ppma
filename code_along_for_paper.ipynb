# Getting the overall amino acid percentages and Adding to DataFrame

```python
#concatone the sequences in each antibody
sequence_concat = {}
i = 0
for thing in list(df['TPP']):
    sequence_concat[f'{thing}']=list(df['Sequence 1'])[i]+ list(df['Sequence 2'])[i]+ list(df['Sequence 3'])[i]+ list(df['Sequence 4'])[i]
    i+=1
```

```python
concatonated_sequences_dataframe = pd.DataFrame(sequence_concat, index = [0])
```

```python
#adding the concatonated sequences to the dataframe
df['concat_seq'] = list(sequence_concat.values())
```

```python
#list of the amino acids
amino_acids = ['A','C','D','E','F','G','H','I','K','L','M','N','P','Q','R','S','T','V','W','Y']
```

```python
#getting percentages of each amino acid in each concatonated sequnce
percentages = {}
for thing in list(df['concat_seq']):
    for letter in amino_acids:
        num_times = 0
        for amino in thing:
            if amino == letter:
                num_times+=1
            else:
                continue
        if letter in percentages:
            percentages[f'{letter}'].append(num_times/len(thing)*100)
        else:
            percentages[f'{letter}'] = [num_times/len(thing)*100]
```

```python
#adding the percent of each amino acid to the dataframe
for thing in list(percentages.keys()):
    df[f'{thing}%'] = percentages[f'{thing}']
```

# Creating Germline and Germline Pairs Columns

```python
#sort the pairs
for thing in list(set(list(list(df['VH/VL Pair #1'])+
                           list(df['VH/VL Pair #2'])))):
    new_thing=' & '.join(sorted(thing.split(' & ')))
    df = df.replace(f'{thing}',new_thing)
```

```python
# Extract individual germlines
germlines_concat = list(df['V-germline (simplified) 1'])+list(df['V-germline (simplified) 2'])+list(df['V-germline (simplified) 3'])+list(df['V-germline (simplified) 4'])
germlines = list(set(germlines_concat))
```

```python
# Extract germline pairs
germline_pairs_concat = list(df['VH/VL Pair #1'])+list(df['VH/VL Pair #2'])
pairs = list(set(germline_pairs_concat))
```

```python
#making a list of all zeroes to have the right amount of zeroes for a column for germlines
zeroes= []
for thing in list(df['V-germline (simplified) 1']):
    zeroes.append(0)
```

```python
#making a column for every individual germline and filling the rows up with zeroes
for thing in germlines:
    df[f'{thing}']= zeroes
```

```python
#for each germline adding in +1 or +2 or however many +'s' for that antibody
for i in list(range(0,len(list(df['Format'])))):
    germlines_=list(df.iloc[i][['V-germline (simplified) 1','V-germline (simplified) 2',
               'V-germline (simplified) 3','V-germline (simplified) 4']])
    for thing in germlines_:
        try:
            if thing in list(df.columns):
                df.at[i,f'{thing}'] = df.iloc[i][f'{thing}']+1
            #this else is for germlines that have not yet been used
            else:
                df[f'{thing}']=zeroes
                df.at[i,f'{thing}'] = df.iloc[i][f'{thing}']+1  
        except:
            print(f'This germline was not added to a list:{thing}')
```

```python
#making a column for every germline pair and filling the rows up with zeroes
for thing in pairs:
    df[f'{thing}']= zeroes
```

```python
#for each germline pair adding in +1 or +2 or however many +'s' for that antibody
for i in list(range(0,len(list(df['Format'])))):
    pairing=list(df.iloc[i][['VH/VL Pair #1','VH/VL Pair #2']])
    for thing in pairing:
        try:
            if thing in list(df.columns):
                df.at[i,f'{thing}'] = df.iloc[i][f'{thing}']+1
            #this else is for pairings that have not yet been used yet
            else:
                df[f'{thing}']=zeroes
                df.at[i,f'{thing}'] = df.iloc[i][f'{thing}']+1 
        except:
            print(f'this pair did not get added:{thing}')
```

# Amino Acid Percentages From Kabat Definition Regions

For this part, you will need to use your own method of obtaining the individual frameworks and CDR components. We use a seaparate csv here that has this information in the 'Martin' column

```python
df2 = pd.read_csv('kabat_region_data.csv', sep = ',')
```

```python
#This function outputs a dictionary where the key is a TPP name, and the values are the propper KABAT region splitting of the sequence
def make_dict_sequences (df,FR1, Sequence,Martin):
    amino_perc_dict = {}
    for row_num in list(range(0,len(list(df['TPP'])))):
        #if the sequence is a light chain
        if 'IGLV' in df.loc[row_num][f'{FR1}'] or 'IGKV' in df.loc[row_num][f'{FR1}']:
            amino_perc_dict[df.loc[row_num]['TPP']]=[]
            counter = 0
            for thing in list(df[f'{Martin}'])[row_num].split('|'):
                if thing != '24':
                    counter+=1
                else:
                    break
            amino_perc_dict[df.loc[row_num]['TPP']].append(df[f'{Sequence}'][row_num][0:counter])

            counter2 = 0
            for thing in list(df[f'{Martin}'])[row_num].split('|'):
                if thing != '35':
                    counter2+=1
                else:
                    break
            amino_perc_dict[df.loc[row_num]['TPP']].append(df[f'{Sequence}'][row_num][counter:counter2])

            counter3 = 0
            for thing in list(df[f'{Martin}'])[row_num].split('|'):
                if thing != '50':
                    counter3+=1
                else:
                    break
            amino_perc_dict[df.loc[row_num]['TPP']].append(df[f'{Sequence}'][row_num][counter2:counter3])

            counter4 = 0
            for thing in list(df[f'{Martin}'])[row_num].split('|'):
                if thing != '57':
                    counter4+=1
                else:
                    break
            amino_perc_dict[df.loc[row_num]['TPP']].append(df[f'{Sequence}'][row_num][counter3:counter4])

            counter5 = 0
            for thing in list(df[f'{Martin}'])[row_num].split('|'):
                if thing != '89':
                    counter5+=1
                else:
                    break
            amino_perc_dict[df.loc[row_num]['TPP']].append(df[f'{Sequence}'][row_num][counter4:counter5])

            counter6 = 0
            for thing in list(df[f'{Martin}'])[row_num].split('|'):
                if thing != '98':
                    counter6+=1
                else:
                    break
            amino_perc_dict[df.loc[row_num]['TPP']].append(df[f'{Sequence}'][row_num][counter5:counter6])

            counter7 = 0
            for thing in list(df[f'{Martin}'])[row_num].split('|'):
                if thing != '107':
                    counter7+=1
                elif thing == '107':
                    counter7+=1
                    #have to add one more at the end becuase need including 107
            amino_perc_dict[df.loc[row_num]['TPP']].append(df[f'{Sequence}'][row_num][counter6:counter7])
        #if the sequence is a heavy chain
        elif 'VH' in df.loc[row_num][f'{FR1}'] or 'IGHV' in df.loc[row_num][f'{FR1}']:
            amino_perc_dict[df.loc[row_num]['TPP']]=[]
            counter = 0
            for thing in list(df[f'{Martin}'])[row_num].split('|'):
                if thing != '31':
                    counter+=1
                else:
                    break
            amino_perc_dict[df.loc[row_num]['TPP']].append(df[f'{Sequence}'][row_num][0:counter])

            counter2 = 0
            for thing in list(df[f'{Martin}'])[row_num].split('|'):
                if thing != '36':
                    counter2+=1
                else:
                    break
            amino_perc_dict[df.loc[row_num]['TPP']].append(df[f'{Sequence}'][row_num][counter:counter2])

            counter3 = 0
            for thing in list(df[f'{Martin}'])[row_num].split('|'):
                if thing != '50':
                    counter3+=1
                else:
                    break
            amino_perc_dict[df.loc[row_num]['TPP']].append(df[f'{Sequence}'][row_num][counter2:counter3])

            counter4 = 0
            for thing in list(df[f'{Martin}'])[row_num].split('|'):
                if thing != '66':
                    counter4+=1
                else:
                    break
            amino_perc_dict[df.loc[row_num]['TPP']].append(df[f'{Sequence}'][row_num][counter3:counter4])

            counter5 = 0
            for thing in list(df[f'{Martin}'])[row_num].split('|'):
                if thing != '95':
                    counter5+=1
                else:
                    break
            amino_perc_dict[df.loc[row_num]['TPP']].append(df[f'{Sequence}'][row_num][counter4:counter5])

            counter6 = 0
            for thing in list(df[f'{Martin}'])[row_num].split('|'):
                if thing != '103':
                    counter6+=1
                else:
                    break
            amino_perc_dict[df.loc[row_num]['TPP']].append(df[f'{Sequence}'][row_num][counter5:counter6])

            counter7 = 0
            for thing in list(df[f'{Martin}'])[row_num].split('|'):
                if thing != '113':
                    counter7+=1
                elif thing == '113':
                    counter7+=1
                    #I have to add one more at the end becuase I need including 107
            amino_perc_dict[df.loc[row_num]['TPP']].append(df[f'{Sequence}'][row_num][counter6:counter7])

    return amino_perc_dict
```

```python
sqs1=make_dict_sequences (df2,'germline 1', 'Original Sequence 1','Martin_s1')
sqs2=make_dict_sequences (df2,'germline 2', 'Original Sequence 2','Martin_s2')
sqs3=make_dict_sequences (df2,'germline 3', 'Original Sequence 3','Martin_s3')
sqs4=make_dict_sequences (df2,'germline 4', 'Original Sequence 4','Martin_s4')
```

```python
#combining all of the chain split ups into 1 dictionary
master_dict = {}
for tpp in list(df2['TPP']):
    master_dict[f'{tpp}']=[sqs1[f'{tpp}'],sqs2[f'{tpp}'],sqs3[f'{tpp}'],sqs4[f'{tpp}']]
```

```python
dict_to_df = {}

for thing in amino_acids:
    dict_to_df[f'{thing}']=[]


for thing in list(master_dict.keys()):
    spec_seq=master_dict[f'{thing}'][0][0]
    for amino in amino_acids:
        dict_to_df[f'{amino}'].append(spec_seq.count(f'{amino}')*100/len(spec_seq))
```

```python
#this is very important, makes dictionary with amino acid percentage
dict_to_df = {}
for num in list(range(0,28)):
    for thing in amino_acids:
        dict_to_df[f'{thing}{num}']=[]
i = -1        
for digit in [0,1,2,3]:
    for value in [0,1,2,3,4,5,6]:
        i+=1
        for thing in list(master_dict.keys()):
            spec_seq=master_dict[f'{thing}'][digit][value]
            for amino in amino_acids:
                dict_to_df[f'{amino}{i}'].append(spec_seq.count(f'{amino}')*100/len(spec_seq))
```
